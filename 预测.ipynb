```python
loss=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/single/record/validate_loss.npy")
acc=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/single/record/validate_accuracy.npy")
miou=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/single/record/validate_m_iou.npy")
dice=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/single/record/validate_dice.npy")

loss1=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/multi/record/validate_loss.npy")
acc1=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/multi/record/validate_accuracy.npy")
miou1=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/multi/record/validate_m_iou.npy")
dice1=np.load("/home/mzhang/project_ASC/journal/ascnet_gpu/multi/record/validate_dice.npy")
print(loss.shape)
print(acc.shape)
print(miou.shape)
start=1


fig=plt.figure(figsize=(18,6))
epoch=range(start,start+len(loss)*100,100)

ax1=fig.add_subplot(1,3,1)
plt.plot(epoch,loss,label="loss")
plt.plot(epoch,loss1,label="loss-multi")
plt.legend(loc="best")

ax1=fig.add_subplot(1,3,2)
plt.plot(epoch,miou,label="acc")
plt.plot(epoch,miou1,label="acc-multi")
plt.legend(loc="best")

ax1=fig.add_subplot(1,3,3)
plt.plot(epoch,dice,label="dice")
plt.plot(epoch,dice1,label="dice-multi")
plt.legend(loc="best")

plt.show()

print(start+200*(np.argmax(acc)))
print(max(acc))
print(miou[np.argmax(acc)])

print(start+200*(np.argmax(miou)))
print(max(miou))
print(dice[np.argmax(miou)])

print(start+200*(np.argmax(dice)))
print(max(dice))
print(acc[np.argmax(dice)])

print(start+200*(np.argmax(dice1)))
print(max(dice1))
print(acc[np.argmax(dice1)])
```
